En este código se va a generar los datos para obtener los datos de los jugadores

```python
import pandas as pd
import numpy as np
from pymongo import MongoClient
import re
```

```python
mongo_url="mongodb+srv://Prueba:Prueba@eurocup-2020.zgna6.mongodb.net/test"

client = MongoClient(mongo_url)
db = client.get_database("EUROCUP-2020")
match_db= db.Matches
```

```python
countries=["England","Italy","Denmark","Spain","Ukraine","Czech Republic","Belgium","Switzerland","Sweden","Germany","France","Croatia","Portugal","Netherlands","Austria","Wales","Hungary","Poland","Slovakia","Scotland","Finland","Russia","North Macedonia","Turkey"]
```

```python
lineup={}
player=[]
count=0

for country in countries: #se itera sobre cada país
    lineup={}

    for match in list(match_db.find({})): #iteración sobre cada partido
    
        if country== match['team_name_home']: #si el país es el equipo que juega en casa se ven los eventos que lleven la etiqueta "casa"
            lineup['country']=country
            
            for evento in match['events_list']:
                if evento['event_team']=='home':
                    
                    if evento['event_type']=='Yellow card': #se comprueban las amarillas y las rojas
                        if len(evento ['action_player_1'].split(" "))==4:
                            jugad=evento ['action_player_1'].split(" ")[-2]
                        elif len(evento ['action_player_1'].split(" "))==6:
                            jugad=" ".join(evento ['action_player_1'].split(" ")[3,5])

                        if jugad not in lineup.keys(): #si el jugador no existe en el diccionario, se añade el nombre, el número 0 y lo que haya pasado
                            player.append(0)
                            player.append(f"Yellow {match['stage']}")
                            lineup[jugad]=player
                            player=[]
                        else:
                            for jug,dat in lineup.items(): #en el caso en el que exista el jugador sólo se añade el evento
                                if jug==jugad:
                                    player=dat
                                    break
                            player.append(f"Yellow {match['stage']}")
                            lineup[jugad]=player
                            player=[]
                    
                    elif evento['event_type']=='Red card': 
                        
                        if len(evento ['action_player_1'].split(" "))==4:
                            jugad=evento ['action_player_1'].split(" ")[-2]
                        elif len(evento ['action_player_1'].split(" "))==6:
                            jugad=" ".join(evento ['action_player_1'].split(" ")[3,5])

                        if jugad not in lineup.keys():
                            player.append(0)
                            player.append('Red')
                            lineup[jugad]=player
                            player=[]
                        else:
                            for jug,dat in lineup.items():
                                if jug==jugad:
                                    player=dat
                                    break
                            player.append(f"Red {match['stage']}")
                            lineup[jugad]=player
                            player=[]
                    
                    elif evento['event_type']=='Goal': #se indica el partido en el que se marca el gol
                        if len(evento ['action_player_1'].split(" "))==4:
                            jugad=evento ['action_player_1'].split(" ")[-2]
                        elif len(evento ['action_player_1'].split(" "))==6:
                            jugad=" ".join(evento ['action_player_1'].split(" ")[3,5])

                        if jugad not in lineup.keys():
                            player.append(0)
                            player.append(f"Goal {match['stage']}")
                            lineup[jugad]=player
                            player=[]
                        else:
                            for jug,dat in lineup.items():
                                if jug==jugad:
                                    player=dat
                                    break
                            player.append(f"Goal {match['stage']}")
                            lineup[jugad]=player
                            player=[]        
                            
            for jugador in match['lineup_home']: #se corrigen los números de los jugadores y se añaden los que no han sido recogidos en un evento
                #los cambios no se tienen en cuenta para la alineación de salida, así que se añaden los jugadores y luego se les asigna el número
                if jugador['Player_Name'] not in lineup.keys():
                    player.append(jugador['Player_Number'])
                    lineup[jugador['Player_Name']]=player
                    player=[]
                if jugador['Player_Name'] in lineup.keys():
                    for i,j in lineup.items():
                        if jugador['Player_Name']==i and j[0]==0:
                            j[0]=jugador['Player_Number']

                
        elif country== match['team_name_away']:#si el país es el equipo que juega fuera de casa se ven los eventos que lleven la etiqueta "fuera"
            lineup['country']=country
                        
            for evento in match['events_list']:
                if evento['event_team']=='away':

                    if evento['event_type']=='Yellow card': 

                        if len(evento ['action_player_1'].split(" "))==4:
                            jugad=evento ['action_player_1'].split(" ")[-2]
                        elif len(evento ['action_player_1'].split(" "))==6:
                            jugad=" ".join(evento ['action_player_1'].split(" ")[3,5])

                        if jugad not in lineup.keys():
                            player.append(0)
                            player.append(f"Yellow {match['stage']}")
                            lineup[jugad]=player

                            player=[]
                        else:
                            for jug,dat in lineup.items():
                                if jug==jugad:
                                    player=dat
                                    break
                            player.append(f"Yellow {match['stage']}")
                            lineup[jugad]=player
                            player=[]
                    
                    elif evento['event_type']=='Red card': 

                        if len(evento ['action_player_1'].split(" "))==4:
                            jugad=evento ['action_player_1'].split(" ")[-2]
                        elif len(evento ['action_player_1'].split(" "))==6:
                            jugad=" ".join(evento ['action_player_1'].split(" ")[3,5])

                        if jugad not in lineup.keys():
                            player.append(0)
                            player.append(f"Red {match['stage']}")
                            lineup[jugad]=player
                            player=[]
                        else:
                            for jug,dat in lineup.items():
                                if jug==jugad:
                                    player=dat
                                    break
                            player.append(f"Red {match['stage']}")
                            lineup[jugad]=player
                            player=[]
                    
                    elif evento['event_type']=='Goal': 

                        if len(evento ['action_player_1'].split(" "))==4:
                            jugad=evento ['action_player_1'].split(" ")[-2]
                        elif len(evento ['action_player_1'].split(" "))==6:
                            jugad=" ".join(evento ['action_player_1'].split(" ")[3,5])
                            
                        if jugad not in lineup.keys():
                            player.append(0)
                            player.append(f"Goal {match['stage']}")
                            lineup[jugad]=player
                            player=[]
                        else:
                            for jug,dat in lineup.items():
                                if jug==jugad:
                                    player=dat
                                    break
                            player.append(f"Goal {match['stage']}")
                            lineup[jugad]=player
                            player=[]        
            
            for jugador in match['lineup_away']:

                if jugador['Player_Name'] not in lineup.keys():
                    player.append(jugador['Player_Number'])
                    lineup[jugador['Player_Name']]=player
                    player=[]
                if jugador['Player_Name'] in lineup.keys():
                    for i,j in lineup.items():
                        if jugador['Player_Name']==i and j[0]==0:
                            j[0]=jugador['Player_Number']

#    for i,j in lineup.items():
#        print(i, j)
#    print('----------------------------------')
    players_db= db.Players
    players_db.insert_one(lineup)

```
