En este c√≥digo se va a generar los datos para obtener los datos de los partidos por equipo

```python
import pandas as pd
import numpy as np
from pymongo import MongoClient
```

```python
mongo_url="mongodb+srv://Prueba:Prueba@eurocup-2020.zgna6.mongodb.net/test"

client = MongoClient(mongo_url)
db = client.get_database("EUROCUP-2020")
match_db= db.Matches
```

```python
countries=["England","Italy","Denmark","Spain","Ukraine","Czech Republic","Belgium","Switzerland","Sweden","Germany","France","Croatia","Portugal","Netherlands","Austria","Wales","Hungary","Poland","Slovakia","Scotland","Finland","Russia","North Macedonia","Turkey"]
```

```python
results={}
count=0
    
for country in countries:
    results={}
    results['country']=country
    for match in list(match_db.find({})):
    
        if country== match['team_name_home']:
            results[match['stage']]=match['date']
            
            #goles del equipo
            if match['team_home_score']=='True':
                results[f"{match['stage']}_goals"]=1
            elif match['team_home_score']=='False':
                results[f"{match['stage']}_goals"]=0
            else:
                results[f"{match['stage']}_goals"]=match['team_home_score']
            
            #penaltis del equipo
            if match['pens']=='False':
                results[f"{match['stage']}_pens"]=0#match['stage']
            else:
                if match['pens_home_score']=='True':
                    results[f"{match['stage']}_pens"]=1
                else:
                    results[f"{match['stage']}_pens"]=match['pens_home_score']


            #nombre del rival
            results[f"{match['stage']}_rival"]=match['team_name_away']

            #goles del rival
            if match['team_away_score']=='True':
                results[f"{match['stage']}_rival_goals"]=1
            elif match['team_away_score']=='False':
                results[f"{match['stage']}_rival_goals"]=0
            else:
                results[f"{match['stage']}_rival_goals"]=match['team_away_score']

            #penaltis del rival
            if match['pens']=='False':
                results[f"{match['stage']}_rival_pens"]=0#match['stage']
            else:
                if match['pens_away_score']=='True':
                    results[f"{match['stage']}_rival_pens"]=1
                else:
                    results[f"{match['stage']}_rival_pens"]=match['pens_away_score']
            
            

            
        elif country== match['team_name_away']:
            results[match['stage']]=match['date']

            #goles del equipo
            if match['team_away_score']=='True':
                results[f"{match['stage']}_goals"]=1
            elif match['team_away_score']=='False':
                results[f"{match['stage']}_goals"]=0
            else:
                results[f"{match['stage']}_goals"]=match['team_away_score']

            #penaltis del equipo
            if match['pens']=='False':
                results[f"{match['stage']}_pens"]=0
            else:
                if match['pens_away_score']=='True':
                    results[f"{match['stage']}_pens"]=1
                else:
                    results[f"{match['stage']}_pens"]=match['pens_away_score']

            #nombre del rival
            results[f"{match['stage']}_rival"]=match['team_name_home']

            #goles del rival
            if match['team_home_score']=='True':
                results[f"{match['stage']}_rival_goals"]=1
            elif match['team_home_score']=='False':
                results[f"{match['stage']}_rival_goals"]=0
            else:
                results[f"{match['stage']}_rival_goals"]=match['team_home_score']

            #penaltis del rival
            if match['pens']=='False':
                results[f"{match['stage']}_rival_pens"]=0#match['stage']
            else:
                if match['pens_home_score']=='True':
                    results[f"{match['stage']}_rival_pens"]=1
                else:
                    results[f"{match['stage']}_rival_pens"]=match['pens_home_score']

        else:
            pass 
    
    goals_db= db.Teams
    goals_db.insert_one(results)
```
